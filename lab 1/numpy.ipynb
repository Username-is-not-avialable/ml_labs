<h1>Задача 1. Дан набор из $p$ матриц размерностью $(n, n)$ и $p$ векторов размерностью $(n, 1)$, найти сумму произведений матриц на векторы. Написать тесты для кода

```python
def sum_prod(X, V):
    '''
    X - матрицы (n, n)
    V - векторы (n, 1)
    Гарантируется, что len(X) == len(V)
    '''
    return sum([np.dot(x,y) for x,y in zip(X,V)])

A = np.array([[1,2],[3,4]])
B = np.array([[1,1],[1,1]])
a = np.array([1,2])
b = np.array([1,1])
X = [A,B]
V = [a,b]
answer = np.array([7, 13])

sum1 = sum_prod(X,V)
print(sum1)
if np.array_equal(sum1, answer) : print ("Test passed!")
```

```python
def binarize(M, threshold=0.5):
    return np.where(M > threshold, 1, 0)

M1 = np.array([[1,2],[-1,-4]])
M2 = np.array([[-1,-2],[-1,-4]])
M3 = np.array([[1,4],[9,11]])

ANS1 = np.array([[1,1],[0,0]])
ANS2 = np.array([[0,0],[0,0]])
ANS3 = np.array([[0,0],[1,1]])

if np.array_equal(binarize(M1), ANS1): print("Test1 passed")
if np.array_equal(binarize(M2), ANS2): print("Test2 passed")
if np.array_equal(binarize(M3, 4), ANS3): print("Test3 passed")
```

```python
def unique_rows(mat):
    return np.apply_along_axis(set, axis = 1, arr = mat)


def unique_columns(mat):
    return np.apply_along_axis(set, axis = 0, arr = mat)


M1 = np.array([[1,2],[3,3]])
M2 = np.array([[1,2,2,2,5,6],[3,3,3,3,5,6]])
print(unique_rows(M1), "уникальные элементы строк матрицы M1")
print(unique_columns(M1), "уникальные элементы колонок матрицы M1")
print(unique_rows(M2), "уникальные элементы строк матрицы M2")
print(unique_columns(M2), "уникальные элементы колонок матрицы M2")
```

<h1>Задача 7. Дан некий временной ряд. Для данного ряда нужно найти его: математическое ожидание, дисперсию, СКО, найти все локальные максимумы и минимумы (локальный максимум - это точка, которая больше своих соседних точек, а локальный минимум - это точка, которая меньше своих соседей), а также вычислить для данного ряда другой ряд, получаемый методом скользящего среднего с размером окна $p$.
<h1>Примечание: метод скользящего среднего подразумевает нахождение среднего из подмножетсва ряда размером $p$

```python
time_series = np.array([1,2,3,2,5,6,7,5,4,5,3])

math_exp = np.mean(time_series)
variance = np.var(time_series)
std = np.std(time_series)
local_maxima = [time_series[i] for i in range(1, len(time_series) - 1) if time_series[i-1] < time_series[i] > time_series[i+1]]
local_minima = [time_series[i] for i in range(1, len(time_series) - 1) if time_series[i-1] > time_series[i] < time_series[i+1]]

print(f"Мат. ождание: {math_exp}")
print(f"Дисперсия: {variance}")
print(f"СКО: {std}")
print(f"Локальне максимумы: {local_maxima}")
print(f"Локальне минимумы: {local_minima}")
```
