#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def get_vowels_count(st):
  st = st.lower()
  return st.count("a")+st.count("e")+st.count("i")+st.count("o")+st.count("u")
def test(st, vowels):
  print(f"Test on \"{st}\" passed") if get_vowels_count(st) == vowels else print(f"Test on \"{st}\" failed")
test("dOg", 1)
test("unicorn", 3)
test("brrrr", 0)
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_every_symbol_uniq(st):
  return True if len(st) == len(set(st)) else False
print(is_every_symbol_uniq("asdf"))
print(is_every_symbol_uniq("aasdf"))
print(is_every_symbol_uniq(""))
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_non_zero_bits(n):
  return bin(n).count("1")


def test(num, ones):
  print(f"Test on number {num} passed") if count_non_zero_bits(num) == ones else print(f"Test on number {num} failed")


test(0,0)
test(1,1)
test(2,1)
test(3,2)
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
from functools import reduce


def magic(n,count = 0):
  if n < 10 : return 0
  prod = reduce(lambda x,y : x*y, [int(digit) for digit in str(n)])
  return magic(prod, count + 1) + 1

def test(num, ans):
  print(f"Test on number {num} passed") if magic(num) == ans else print(f"Test on number {num} failed")


test(39, 3)
test(4, 0)
test(999, 4)
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if (len(pred) != len(true)): return "Incorrect input: vectors have different length"

    squared_differences = list(map(lambda x,y : (x - y)**2,pred, true)) # квадраты разностей
    mean_squared = sum(squared_differences) / len(squared_differences) # среднее квадратов разностей
    standart_deviation = mean_squared**0.5

    return standart_deviation


vector1 = [1, 2, 3, 4, 5]
vector2 = [2, 3, 4, 5, 6]
print(mse(vector1, vector2) == 1.0)
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
from math import sqrt
def prime_factorize(n):
  factors = {}
  for i in range (2, int(sqrt(n))):
    while n % i == 0:
      if i not in factors.keys(): factors[i] = 1
      else: factors[i] += 1
      n //= i

  if n > 1:
    if n not in factors.keys(): factors[n] = 1
    else: factors[n] += 1

  factors_str = ""
  for item in factors.items():
    if item[1] != 1:
      factors_str += f"({item[0]}**{item[1]})"
    else:
      factors_str += f"({item[0]})"

  return factors_str

print(prime_factorize(86240))
print(prime_factorize(10))
print(prime_factorize(20))
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
  layer = 1
  while number > 0:
    number -= layer**2
    if number == 0: return layer
    if number > 0:
      layer += 1
      continue
  return "It is impossible"


def test(num, ans):
  print(f"Test on number {num} passed") if pyramid(num) == ans else print(f"Test on number {num} failed")


test(1,1)
test(2, "It is impossible")
test(5,2)
test(7, "It is impossible")
test(14,3)
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
  num = [int(dig) for dig in str(n)]
  if len(num) % 2 == 0:
    sum1 = sum(num[:len(num)//2-1])
    sum2 = sum(num[len(num)//2+1:])
    return sum1 == sum2
  sum1 = sum(num[:len(num)//2])
  sum2 = sum(num[len(num)//2+1:])
  return sum1 == sum2


def test(num, ans):
  print(f"Test on number {num} passed") if is_balanced(num) == ans else print(f"Test on number {num} failed")

test(1221, True)
test(1234321, True)
test(23141, True)
test(113, False)
test(1133, False)
```
